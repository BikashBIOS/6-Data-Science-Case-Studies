### TASK 3 : Data Cleansing

1. Take all String values Columns and convert them into binaries (encode them) using OneHotEncoder (from scikitlearn). We do this because we can analyse the data with only binary.

2. Mix all the numerical columns and encoded string columns in one df.(one thing to note - don't include your target column in the DF. In our case, the target column in Attrition, so we will drop the Attrition column from our mix df). (Target column is basically, the column that we need to predict).

3. Next step is to scale the data - mostly we need to adjust all the columns data in such a way that there is no much difference between one column values and another column values. (basically the range of the integers between all the columns should be minimal).

4. Now take y as the target column which is Attrition.


Steps we proceeded :

1. x_cat = categorical columns (string columns) - encoded by OneHotEncoder.
2. x_num = numerical columns (excluding the target - Attrition).
3. x_all = concat the x_cat and x_num
4. x = scale the x_all - by MinMaxScaler
5. y = target column

### TASK 7 : Train & Evaluate a Deep Learning Model

Here, we will train the model for Binary classification.

Steps :

1. Use **Tensorflow** because it integrates **Keras**, which is a high level API that makes it easy to train and build Neural networks.
2. Keras has different models - here we use **Sequential** which allows us to create a linear stack of layers for our neural network architecture.
3. Then we create layers in a sequential manner. Each layer is added over the previous one - forming a chain of layers.
4. 1st Layer (Input layer) - 500 neurons, Activation function is **ReLU** (Rectified Linear Unit) - *, input data is 50 features (as there are almost 50 columns in our dataset) (input data will only be present in the first layer).
5. 2nd & 3rd layers - to make it more dense.
6. 4th Layer (Outer layer) - used for binary classification - therefore we use one 1 neuron and Activation function is **Sigmoid** - sigmoid function returns a value between 0 and 1.

* ReLU is used for Non-linearity and complex relationship in models.
Without non-linearity, no matter how many layers a neural network has, it would behave like a single-layer linear model.
ReLU is typically used in the hidden layers of neural networks to introduce non-linearity and help the network learn complex patterns.



```python
import tensorflow as tf

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units = 500, activation = 'relu', input_shape =(50, )))
model.add(tf.keras.layers.Dense(units = 500, activation = 'relu'))
model.add(tf.keras.layers.Dense(units = 500, activation = 'relu'))
model.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))
```

```python
model.summary()
```

7. Now, we will compile our model specifying how the model should be configured for training.
8. Optimizer is an algo that adjusts the weights of NN to minimize Loss function during training. Here we use Adam - Adaptive Moment Estimation - maintains an adaptive learning rate for each parameter by storing the first and second moments of gradients.

9. loss() measures how well the model predicts the true labels. During training, optimizer tries to minimize this loss function.
10. Binary cross entropy - calculates the cross-entropy loss between true labels and predicted probabilities.
11. metrics used to evaluate the performance of a model with Accuracy being the most commonly used.
Accuracy = No. of Correct Predictions / Total No. of Predictions.

```python
model.compile(optimizer = 'Adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

12. Now, we will train the data using fit(). Epochs - one complete pass through the whole dataset. (here 100 epochs means it will go through the whole dataset 100 times). Batch size - no. of samples that will go through network at one time. (Smaller batch sizes provide more updates per epoch but are more computationally expensive. Larger batch sizes make the training process faster but may converge slower or get stuck in local minima).
13. Summary - The model is trained 100 times over the dataset in which the weights are updated after processing every 50 samples. Then the training history - loss & metric for each epoch - is stored in epoch_hist.

```python
epochs_hist = model.fit(x_train, y_train, epochs = 100, batch_size = 50)
```

14. Then predict the loss and accuracy using x and y data. Plot it on the graph for loss and accuracy. As you can see from graph, loss decreases gradually and Accuracy increase gradually.
15. Proceed with Confustion matrix for y.

```python
y_pred = model.predict(x_test)
y_pred =(y_pred > 0.5)
y_pred
```

```python
plt.plot(epochs_hist.history['loss'])
plt.title('Model Loss Progress During Training')
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.legend(['Training Loss'])
```

```python
plt.plot(epochs_hist.history['accuracy'])
plt.title('Model Accuracy Progress During Training')
plt.xlabel('Epoch')
plt.ylabel('Training Accuracy')
plt.legend(['Training Accuracy'])
```

```python
cm = confusion_matrix(y_pred, y_test)
sns.heatmap(cm, annot=True)
```
